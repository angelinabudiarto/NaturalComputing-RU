```python
# Imports
import math
import matplotlib.pyplot as plt
```

```python
# Definintions of methods used to answer the questions
def calculate_probability(c_amount, p_competence):
    
    majority = math.floor(c_amount/2 + 1)
    
    probability = 0
    for i in range(majority, c_amount+1):
        probability += math.comb(c_amount, i) * p_competence**i * (1-p_competence)**(c_amount-i)
    
    return probability
```

```python
# Question 2b
print(calculate_probability(31, 0.6))
```

```python
# Question 2c
p2_graph = [calculate_probability(x, 0.2) for x in range(5, 51)]
p35_graph = [calculate_probability(x, 0.35) for x in range(5, 51)]
p5_graph = [calculate_probability(x, 0.5) for x in range(5, 51)]
p65_graph = [calculate_probability(x, 0.65) for x in range(5, 51)]
p8_graph = [calculate_probability(x, 0.8) for x in range(5, 51)]

plt.figure(figsize=(8,5))
plt.plot(p2_graph, label='p=0.2')
plt.plot(p35_graph, label='p=0.35')
plt.plot(p5_graph, label='p=0.5')
plt.plot(p65_graph, label='p=0.65')
plt.plot(p8_graph, label='p=0.80')
plt.title('Different competence levels on a range of jury sizes (5-50)')
plt.ylabel('Probability of correct majority vote')
plt.xlabel('Jury size')
plt.xticks(range(0, 46, 5), range(5, 51, 5))
plt.legend()
plt.show()
```

```python
# Question 2d
students_graph = [calculate_probability(x, 0.6) for x in range(22, 32)]

plt.figure(figsize=(8,5))
plt.plot(students_graph)
plt.title('Comparing student maj. vote prob. with doctor maj. vote prob.')
plt.ylabel('Probability of correct majority vote')
plt.xlabel('Student group size')
plt.xticks(range(0, 10), range(22, 32))
plt.axhline(y=0.84375, color="black", linestyle="--")
plt.show()
```
