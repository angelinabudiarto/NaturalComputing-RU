```
# part a) 

import numpy as np
import matplotlib.pyplot as plt
import random

l = 100
iters = 1500
p = 1/l

def mutation_prob(x_m, p):
    for index, value in enumerate(x_m):
        if random.random() < p:
            x_m[index] = not x_m[index]
    return x_m     

def genetic_algorithm (l, iters):
    #(1+1)-GA

    #part 1) Randomly generate a bit sequence x_m
    x_m = np.random.randint(2, size=(l,))

    best_fitness = np.sum(x_m) #counting the ones
    best_fitness_list = [best_fitness]

    #part 4) Repeat the process from step (2) with the new x until the goal sequence is reached
    for i in range(iters):
        #part 2) Create a copy of x and invert each of its bits with probability p. Let x_m be the result
        best_fitness_copy = x_m.copy() #copy
        mutation_bits = mutation_prob(best_fitness_copy, p)
        mutation_bit_fitness = np.sum(mutation_bits) 

        #part 3) If x_m is closer to the goal sequence than x then replace x with x_m    
        if mutation_bit_fitness >= best_fitness:
            x_m = mutation_bits
            best_fitness = mutation_bit_fitness
        
        best_fitness_list.append(best_fitness)

        if (mutation_bit_fitness == l):
            print("Optimum solution found with fitness ", best_fitness, " at iteration", i)
            break

    if (i == (iters-1)):
        print("No optimum solution found, best fitness is ", best_fitness)

    return [x_m, best_fitness, best_fitness_list]


results = genetic_algorithm(l, iters)
  
plt.plot(results[2])
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.title('Best fitness found for (1+1)-GA Counting Ones Problem')
plt.show()
```

```
#part b)

for i in range(10):
    results = genetic_algorithm(l, iters)
    legend = "fitness iteration"+str(i)
    plt.plot(results[2], label=legend)
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.title('Best fitness found for (1+1)-GA Counting Ones Problem')
plt.show()
```
